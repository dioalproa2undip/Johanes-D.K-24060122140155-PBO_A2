```python
#Johanes D.K
#NIM : 24060122140155
#soal 3 dan 4
import pandas as pd
#import dataset iris
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
#jenis kolom
names =['sepal-width', 'petal-length', 'petal-width', 'class']
dataset = pd.read_csv(url, header=None, names=names)
#code untuk mencetak fitur 2 dan 3
fitur_2_dan_3 = dataset[['sepal-width', 'petal-length']]
print(fitur_2_dan_3.head())
```

```python
#plot dataset
import matplotlib.pyplot as plt
#untuk mencetak fitur 2 dan 3
f1= dataset['sepal-width'].values
f2= dataset['petal-length'].values
#plt.scatter(f1[:50], f2[:50], color='red', label='Iris-setosa')
#print hasil
plt.scatter(f1, f2, c='black', s=7)
```

```python
from sklearn.cluster import KMeans
#menentukan jumlah cluster
kmeans=KMeans(n_clusters=3)
#fitting input data
kmeans.fit(fitur_2_dan_3)
#mendapatkan nilai centroidnya
labels = kmeans.predict(fitur_2_dan_3)
#mencetak nilai centroid
c = kmeans.cluster_centers_
print(c)
```

```python
#disini adalah menentukan kluster fitur 2 dan 3 dimana fitur 2 bewarna merah dan fitur 3 berwarna biru 
plt.scatter(f1[:50], f2[:50], color='red', label='Iris-setosa')
plt.scatter(f1[50:100], f2[50:100], color='blue', label='Iris-versicolor')
```

```python
import matplotlib.pyplot as plt
for k in range (1, 10):
  #menentukan jumlah cluster
  kmeans=KMeans(n_clusters=k, random_state=1)
  #fitting input data
  kmeans=kmeans.fit(fitur_2_dan_3)
  #code ini untuk mendapatkan cluster labels
  labels = kmeans.predict(fitur_2_dan_3)
  #code untuk menghitung jumlah jarak setiap sampel dengan cluster centroidnya(SSE)
  interia = kmeans.inertia_
  print(k, interia)
  plt.plot(k, interia, 'bx-')
```

```python
from sklearn.metrics.cluster import silhouette_score
silhouette_score(fitur_2_dan_3, labels)
```
